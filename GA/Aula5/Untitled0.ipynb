```python
def tokenize(texto):
    return texto.lower().split()
```

```python
import math

def calcular_prob_log(classe, palavras):
    log_prob = math.log(contagem_emails[classe] / sum(contagem_emails.values()))
    for palavra in palavras:
        contagem = frequencia_por_classe[classe][palavra]
        prob = (contagem + 1) / (total_palavras[classe] + len(vocabulario))
        log_prob += math.log(prob)
    return log_prob
```

```python
def classificar(email_texto):
    palavras = tokenize(email_texto)
    log_prob_spam = calcular_prob_log("spam", palavras)
    log_prob_nao_spam = calcular_prob_log("nao_spam", palavras)

    if log_prob_spam > log_prob_nao_spam:
        classe = "SPAM"
        prob = math.exp(log_prob_spam - log_prob_spam) / (
            math.exp(log_prob_spam - log_prob_spam) + math.exp(log_prob_nao_spam - log_prob_spam))
    else:
        classe = "NÃO SPAM"
        prob = math.exp(log_prob_nao_spam - log_prob_nao_spam) / (
            math.exp(log_prob_spam - log_prob_nao_spam) + math.exp(log_prob_nao_spam - log_prob_nao_spam))

    return classe, round(prob * 100, 2)
```

```python
# 1. Instalar a biblioteca (caso ainda não esteja instalada)
!pip install -q datasets

# 2. Carregar o dataset Sentiment140 diretamente
from datasets import load_dataset

dataset = load_dataset('sentiment140')

import pandas as pd

df = pd.DataFrame(dataset['train'])
df = df[['text', 'sentiment']]  # 'sentiment': 0 = negativo, 4 = positivo

df.head()

```

```python
from sklearn.utils import shuffle

df = shuffle(df)

df.head(10)
```

```python
for index, row in df.iterrows():
  print(row.sentiment)
```

```python
from collections import defaultdict

vocabulario = set()
frequencia_por_classe = {0: defaultdict(int), 4: defaultdict(int)}
contagem_emails = {0: 0, 4: 0}
total_palavras = {0: 0, 4: 0}
contador = 0
for i, infos in df.iterrows():
  if contador == 10000:
    break
  classe = infos.sentiment
  contagem_emails[classe] += 1
  palavras = tokenize(infos.text)
  contador += 1
  for palavra in palavras:
      vocabulario.add(palavra)
      frequencia_por_classe[classe][palavra] += 1
      total_palavras[classe] += 1
```
